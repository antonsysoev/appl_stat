```r
data <-fread("/content/dib_dataset.csv")
data_analysis <- na.omit(data.frame(data$gdp_per_capita_us_dollars_2017, 
                            data$agricultural_production_index_base_2004_2006_100_2016,
                            data$food_production_index_base_2004_2006_100_2016,
                            data$urban_population_percent_2018,
                            data$sex_ratio_males_per_100_females_2019,
                            data$population_density_2019,
                            data$employment_by_industry_industry_male_and_female_2019,
                            data$life_expectancy_at_birth_for_both_sexes_years_2015,
                            data$forest_cover_thousand_hectares_2016,
                            data$emissions_per_capita_metric_tons_of_carbon_dioxide_2014))
```

Вычисление главных компонентов

```r
res.pca <- prcomp(data_analysis)
```

```r
fviz_eig(res.pca)
```

```r
# Eigenvalues
eig.val <- get_eigenvalue(res.pca)
eig.val
  
# Results for Variables
res.var <- get_pca_var(res.pca)
res.var$coord          # Coordinates
res.var$contrib        # Contributions to the PCs
 
```
